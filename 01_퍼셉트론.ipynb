## 03 í¼ì…‰íŠ¸ë¡  êµ¬í˜„í•˜ê¸°

### 03.01 ê°„ë‹¨í•œ êµ¬í˜„ë¶€í„°

```python
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1*w1 + x2*w2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1
```

```python
print(AND(0, 0))
print(AND(1, 0))
print(AND(1, 0))
print(AND(1, 1))
```

### 03.03 ê°€ì¤‘ì¹˜ì™€ í¸í–¥ êµ¬í˜„í•˜ê¸°

```python
def AND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  tmp = np.sum(x*w) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

- ê°€ì¤‘ì¹˜: ê° ì…ë ¥ ì‹ í˜¸ê°€ ê²°ê³¼ì— ì£¼ëŠ” ì˜í–¥ë ¥ (ì¤‘ìš”ë„)ì„ ì¡°ì ˆí•˜ëŠ” ë§¤ê°œë³€ìˆ˜
- í¸í–¥: ë‰´ëŸ°ì´ ì–¼ë§ˆë‚˜ ì‰½ê²Œ í™œì„±í™”(ê²°ê³¼ë¡œ 1ì„ ì¶œë ¥)í•˜ëŠëƒë¥¼ ì¡°ì •í•˜ëŠ” ë§¤ê°œë³€ìˆ˜
- ë¬¸ë§¥ì— ë”°ë¼ ê°€ì¤‘ì¹˜ì™€ í¸í–¥ì„ ëª¨ë‘ ê°€ì¤‘ì¹˜ë¼ê³  í•  ë•Œë„ ìˆìŒ

```python
def NAND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([-0.5, -0.5])
  b = 0.7
  tmp = np.sum(x*w) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
def OR(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.2
  tmp = np.sum(x*w) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

### 05.02 XOR ê²Œì´íŠ¸ êµ¬í˜„í•˜ê¸°

```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)
  return y
```
